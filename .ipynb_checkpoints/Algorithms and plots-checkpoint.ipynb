```python
import pandas as pd
import numpy as np

from scipy.sparse import csr_matrix
from scipy.sparse.linalg import svds, norm

import matplotlib.pyplot as plt
import seaborn as sns

from tqdm import tqdm
import os
import pickle

from FW_project_functions import (
    
    cost_func_sp,
    gradient_func_sp,
    linear_oracle_sp,
    
    frank_wolfe_sp,
    away_steps_frank_wolfe_sp,
    pairwise_frank_wolfe_sp,
    
    plot_summary,
    plot_comparisons,
    delta_change_plot,
    delta_change_subplot
    
)
```

# Dataset 0: Randomly generated reviews with missing values

## Dataset

```python
dataset = 'random_dataset'

np.random.seed(42)
size=(100,100)
U = 1.*np.random.randint(1,6, size=size)*(np.random.uniform(low=0, high=1, size=size) > 0.6)

U[U==0] = np.nan


U_sp = csr_matrix(np.nan_to_num(U))
mask = csr_matrix(~np.isnan(U))
```

```python
main_kwargs = dict(
    U_sp = U_sp,
    tol = 1e-6,
    max_iter = 200,
    mask = mask
)
```

## Effect of the $\delta$ parameter

```python
fig, axs = plt.subplots(2,2, figsize=(10, 7))
axs[1,1].remove()

fig.suptitle('Effect of parameter $\delta$\n' +
             f"$shape(U)$ = {U.shape}\n" +
             f"Non-null entries: {1 - np.isnan(U).mean():.0%}")

delta_kwargs = dict(
    name = 'Vanilla FW',
    fw_function = frank_wolfe_sp,
    delta_min = 1,
    delta_max = 3,
    kwargs = main_kwargs,
    ax = axs[0,0],
    N = 20,
)
delta_change_subplot(**delta_kwargs)

delta_kwargs = dict(
    name = 'Away step FW',
    fw_function = away_steps_frank_wolfe_sp,
    delta_min = 1,
    delta_max = 3,
    kwargs = main_kwargs,
    ax = axs[0,1],
    N = 20,
)
delta_change_subplot(**delta_kwargs)

delta_kwargs = dict(
    name = 'Pairwise FW',
    fw_function = pairwise_frank_wolfe_sp,
    delta_min = 1,
    delta_max = 3,
    kwargs = main_kwargs,
    ax = axs[1,0],
    N = 20
)
delta_change_subplot(**delta_kwargs)

fig.tight_layout()
plt.show()
```

## Algorithms

```python
kwargs = main_kwargs.copy()
kwargs['delta'] = 1_000
kwargs['disable'] = False

results = dict()
```

### Simple FW

```python
%%time
color = 'green'
linestyle = '-'

X, costs = frank_wolfe_sp(**kwargs)
costs.color = color
costs.linestyle = linestyle
results['Simple'] = dict(
    X = X,
    costs = costs
)

plot_summary('Simple FW', costs, dataset)
```

### Step away FW

```python
%%time
color = 'orange'
linestyle = '-'

X, costs = away_steps_frank_wolfe_sp(**kwargs)
costs.color = color
costs.linestyle = linestyle
results['Away step'] = dict(
    X = X,
    costs = costs
)

plot_summary('Away steps', costs, dataset)
```

### Pairwise FW

```python
%%time
color = 'blue'
linestyle = '--'

X, costs = pairwise_frank_wolfe_sp(**kwargs)
costs.color = color
costs.linestyle = linestyle
results['Pairwise'] = dict(
    X = X,
    costs = costs
)

plot_summary('Pairwise FW',costs, dataset)
```

### Comparisons

```python
plot_comparisons(results, kwargs, dataset, figsize=(9, 4), loc="upper right")
```
